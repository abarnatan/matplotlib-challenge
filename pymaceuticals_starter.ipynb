## Observations and Insights 

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

merged_df = pd.merge(mouse_metadata, study_results, how = "left", on = "Mouse ID")

# Display the data table for preview

merged_df.head()
```

```python
# Checking the number of mice.

len(merged_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

merged_df[merged_df.duplicated(['Mouse ID', 'Timepoint'])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

merged_df.loc[merged_df['Mouse ID'] == 'g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

merged_clean_df = merged_df.loc[merged_df['Mouse ID'] != 'g989']
merged_clean_df
```

```python
# Checking the number of mice in the clean DataFrame.

len(merged_clean_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

clean_data_grp = merged_clean_df.groupby(["Drug Regimen"])

drug_mean= clean_data_grp["Tumor Volume (mm3)"].mean()
drug_med = clean_data_grp["Tumor Volume (mm3)"].median()
drug_var = clean_data_grp["Tumor Volume (mm3)"].var()
drug_std = clean_data_grp["Tumor Volume (mm3)"].std()
drug_sem = clean_data_grp["Tumor Volume (mm3)"].sem()

summary_df = pd.DataFrame({"Mean":drug_mean, "Median":drug_med, "Variance":drug_var, "STDEV":drug_std, "SEM":drug_sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line 

clean_agg = clean_data_grp.agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', st.sem]})
clean_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

drug_list = ('Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin')
drug_data = merged_clean_df.loc[merged_clean_df['Drug Regimen'].isin(drug_list)]

# Start by getting the last (greatest) timepoint for each mouse

tumor_vol_df = drug_data.groupby(drug_data['Mouse ID']).agg({'Timepoint':['max']})
tumor_vol_df.columns = ['Timepoint']

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

tumor_vol_df = tumor_vol_df.reset_index()
final_tumor_vol_df = pd.merge(tumor_vol_df, merged_clean_df, how="left", on=["Mouse ID", "Timepoint"])
final_tumor_vol_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
for drug in drug_list:
    tumor_vol_by_drug = final_tumor_vol_df['Tumor Volume (mm3)'].loc[final_tumor_vol_df['Drug Regimen'] == drug]
    
    # add subset 
    
    tumor_vol_list.append(tumor_vol_by_drug)
    
    # Determine outliers using upper and lower bounds
    
    quartiles = tumor_vol_by_drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'For {drug}, the interquartile range (IQR) is {iqr}')
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'For {drug}, values below {lower_bound} could be outliers')
    print(f'For {drug}, values above {upper_bound} could be outliers\n')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Identify outliers

red_circle = dict(markerfacecolor='r', marker='o')

fig, ax = plt.subplots()
ax.set_title('Final Tumor Volume per Regimen')
ax.set_xticklabels(drugs)
ax.set_ylabel('Tumor Volume (mm3)')
ax.boxplot(tumor_vol_list, flierprops=red_circle)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

# select a mouse treated with Capomulin

cap_mouse = merged_clean_df[['Timepoint', 'Tumor Volume (mm3)']].loc[(merged_clean_df['Drug Regimen'] == 'Capomulin') & (merged_clean_df['Mouse ID']=='f966')]

# plot data

plt.plot(cap_mouse['Timepoint'], cap_mouse['Tumor Volume (mm3)'], marker='o')
plt.title("Capomulin Regimen - Mouse (f966)")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

cap_avgtumor = merged_clean_df.loc[merged_clean_df['Drug Regimen'] == 'Capomulin'].groupby(merged_clean_df['Timepoint']).agg({'Tumor Volume (mm3)':['mean'], 'Weight (g)':['mean']})
cap_avgtumor.columns = ['Average Tumor Volume {mm3}', 'Average Mouse Weight (g)']

avgtumvol = cap_avgtumor['Average Tumor Volume {mm3}']
avgmowgt = cap_avgtumor['Average Mouse Weight (g)']
plt.scatter(avgmowgt,avgtumvol)
plt.title('Avg Tumor Volume vs. Avg Mouse Weight')
plt.xlabel('Mouse Weight')
plt.ylabel('Tumor Volume')
plt.ylim(35,46)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# colculate r value

correlation = st.pearsonr(avgtumvol,avgmowgt)
print(f'The correlation between Average Tumor Volume and Mouse Weight is {round(correlation[0],2)}')

# calculate linear regression

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(avgmowgt,avgtumvol)
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(f'The linear regression equation is {line_eq}')

# plot line with scatter

avgtumvol = cap_avgtumor['Average Tumor Volume {mm3}']
avgmowgt = cap_avgtumor['Average Mouse Weight (g)']
plt.scatter(avgmowgt,avgtumvol)
plt.title('Avg Tumor Volume vs. Avg Mouse Weight')
plt.xlabel('Mouse Weight')
plt.ylabel('Tumor Volume')
plt.ylim(35,46)

# calculate regression values

reg_values = avgmowgt * slope + intercept
plt.plot(avgmowgt, reg_values, "r-")
plt.annotate(line_eq,(19.95,38),fontsize=15,color="red")
plt.show()
```
